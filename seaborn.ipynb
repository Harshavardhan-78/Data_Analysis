# Seabon-Statistical data visualisation

## What is Seaborn?
seaborn is a python library that builts on top of matplotlib that makes the statisticals data more easy and beautiful
## why we need it?
.for creating plots like violin,box,plain(pair)
.automatically handles data from pandas dataframes
## where we use it?
Data analysis and reporting
Exploring relation between variables
Machine learning :for visualizing datasets and visuals

## 2.Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
print(df.shape)
df.head()
```

## 4. Histograms & Countplots
### Histogram

```python
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
```

### Countplot

```python

sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

### 5. Scatter Plots

```python

sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

## 6. Boxplot & Violin Plot




### Boxplot

```python
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
```


 
### Violin plot

```python
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

## 8. Heatmap
### Correlation heatmap

```python
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

## 9. Customization & Themes


```python
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
 
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
"Show trends or changes in mean values across categories."
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='class', y='survived', hue='sex')
plt.title('Passenger Survival by Class and Gender')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='class', y='fare')
plt.title('Fare Distribution by Class')
plt.show()
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='class', palette=['#DC143C', '#3CB371', '#6495ED'])
plt.title('Relationship Between Age and Fare')
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr_matrix = df[['survived', 'age', 'fare']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True, linewidths=0.5, linecolor='white')
plt.title('Correlation Heatmap')
plt.show()
```
